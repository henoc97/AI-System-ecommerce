### Analyse de sentiment


```python
from transformers import pipeline

# Charger un modèle fine-tuné pour l'analyse de sentiment en français
sentiment_analyzer = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Exemple d'avis en français
avis = "Le produit est très mauvais, j'en suis pas du tout satisfait !"

# Analyse de sentiment
resultat = sentiment_analyzer(avis)
print(resultat)  # Affiche le sentiment
```

### Extraction d'Entités Nommées (NER) avec Hugging Face

```python
from transformers import pipeline

# Charger le pipeline de NER avec un modèle adapté pour le français
ner_tagger = pipeline("ner", model="Jean-Baptiste/camembert-ner", aggregation_strategy="simple")

# Exemple de texte à analyser
texte = "Emmanuel Macron est le président de la République française."

# Extraction des entités nommées
resultat = ner_tagger(texte)

# Afficher les résultats
print(resultat)
```
